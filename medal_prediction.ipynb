```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
teams = pd.read_csv('teams.csv')
```

```python
teams
```

```python
teams = teams[['team','country','year','athletes','age','medals','prev_medals']]
```

```python
teams
```

```python
sel = teams.select_dtypes(include=['int64','float64']).columns
```

```python
teams[sel].corr()["medals"]
```

```python
sns.lmplot(x="age",y="medals",data = teams, fit_reg = True , ci=None)
```

```python
teams[teams.isnull().any(axis=1)]
```

```python
teams = teams.dropna()
```

```python
teams
```

```python
train = teams[teams["year"]<2012].copy()
test = teams[teams["year"]>=2012].copy()
```

```python
train.shape
```

```python
test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

x = ["athletes" , "prev_medals"]
y = "medals"

lr.fit(train[x],train[y])
```

```python
predictions = lr.predict(test[x])
```

```python
predictions
```

```python
test["predictions"] = predictions
```

```python
test
```

```python
test.loc[test["predictions"]<0,"predictions"] = 0
```

```python
test["predictions"]=test["predictions"].round()
```

```python
test
```

```python
from sklearn.metrics import mean_absolute_error
error = mean_absolute_error(test["medals"] , test["predictions"] )
```
